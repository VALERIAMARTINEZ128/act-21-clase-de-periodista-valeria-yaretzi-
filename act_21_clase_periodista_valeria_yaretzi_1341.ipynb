```python
# 1. Definimos la Clase (El molde) 🏗️
class Periodista:
    def __init__(self, nombre, medio, especialidad):
        # Definimos los 3 atributos 📌
        self.nombre = nombre
        self.medio = medio
        self.especialidad = especialidad

    # Definimos las 3 funciones (Acciones) 🎬
    def investigar(self):
        print(f"🔍 {self.nombre} está buscando fuentes sobre {self.especialidad}.")

    def redactar(self):
        print(f"⌨️ Escribiendo una nueva nota para el medio '{self.medio}'...")

    def publicar(self):
        print(f"🚀 ¡Atención! La noticia de {self.nombre} ya está disponible en la web.")

# --- 2. Creación de Objetos (Instancias) 👥 ---

# Creamos al Periodista 1 (Objeto 1)
periodista1 = Periodista("Elena Prensa", "Diario Nacional", "Política 🏛️")

# Creamos al Periodista 2 (Objeto 2)
periodista2 = Periodista("Marcos Goles", "Deportes 24/7", "Fútbol ⚽")

# --- 3. Ejecución de las Funciones (Mensajes) ---

print("--- 📄 REPORTE: PERIODISTA 1 ---")
print(f"Nombre: {periodista1.nombre}")
periodista1.investigar()
periodista1.redactar()
periodista1.publicar()

print("\n--- 📄 REPORTE: PERIODISTA 2 ---")
print(f"Nombre: {periodista2.nombre}")
periodista2.investigar()
periodista2.redactar()
periodista2.publicar()
print("programa termiando por valeria yaretzi NL = 32")
```
